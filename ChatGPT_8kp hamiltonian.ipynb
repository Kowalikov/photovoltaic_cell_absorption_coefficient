```python
def eight_band_kp_Hamiltonian(k):
  # Define the 8x8 Hamiltonian matrix
  H = np.zeros((8,8), dtype=np.complex128)

  # Define the material parameters
  E_g = 0.812 # Bandgap energy
  m_e = 0.086 # Electron effective mass
  m_h = 0.033 # Hole effective mass
  hbar = 1.0545718e-34 # Reduced Planck constant

  # Define the spin-orbit coupling constants
  gamma_1 = 13.4
  gamma_2 = 4.7
  gamma_3 = 6.0

  # Fill in the diagonal elements of the Hamiltonian
  H[0,0] = E_g/2 + 3*(m_e**2)/(2*(m_e+m_h)**2)
  H[1,1] = E_g/2 + (m_e**2)/(2*(m_e+m_h)**2)
  H[2,2] = E_g/2 - (m_e**2)/(2*(m_e+m_h)**2)
  H[3,3] = E_g/2 - 3*(m_e**2)/(2*(m_e+m_h)**2)
  H[4,4] = -E_g/2 - 3*(m_h**2)/(2*(m_e+m_h)**2)
  H[5,5] = -E_g/2 - (m_h**2)/(2*(m_e+m_h)**2)
  H[6,6] = -E_g/2 + (m_h**2)/(2*(m_e+m_h)**2)
  H[7,7] = -E_g/2 + 3*(m_h**2)/(2*(m_e+m_h)**2)

  # Fill in the off-diagonal elements of the Hamiltonian
  H[0,1] = H[1,0] = (-3*m_e**2)/(4*(m_e+m_h)**2)
  H[0,2] = H[2,0] = (3*m_e**2)/(4*(m_e+m_h)**2)
  H[1,2] = H[2,1] = (-m_e**2)/(2*(m_e+m_h)**2)
  H[0,3] = H[3,0] = (m_e**2)/(4*(m_e+m_h)**2)
  H[1,3] = H[3,1] = (-m_e**2)/(4*(m_e+m_h)**2)
  H[2,3] = H[3,2] = (-m_e**2)/(4*(m_e+m_h)**2)
  H[4,5] = H[5,4] = (-3*m_h**2)/(4*(m_e+m_h)**2)
  H[4,6] = H[6,4] = (3*m_h**2)/(4*(m_e+m_h)**2)
  H[5,6] = H[6,5] = (-m_h**2)/(2*(m_e+m_h)**2)
  H[4,7] = H[7,4] = (m_h**2)/(4*(m_e+m_h)**2)
  H[5,7] = H[7,5] = (-m_h**2)/(4*(m_e+m_h)**2)
  H[6,7] = H[7,6] = (-m_h**2)/(4*(m_e+m_h)**2)

  # Add the spin-orbit coupling terms
  H[0,4] = H[4,0] = -1j*gamma_1*k[2]/2
  H[0,5] = H[5,0] = -1j*gamma_2*k[1]/2
  H[0,6] = H[6,0] = -1j*gamma_3*k[0]/2
  H[1,4] = H[4,1] = 1j*gamma_1*k[1]/2
  H[1,7] = H[7,1] = 1j*gamma_3*k[2]/2
  H[2,4] = H[4,2] = 1j*gamma_1*k[0]/2
  H[2,5] = H[5,2] = -1j*gamma_3*k[2]/2
  H[3,5] = H[5,3] = 1j*gamma_2*k[0]/2
  H[3,6] = H[6,3] = 1j*gamma_3*k[1]/2
  H[3,7] = H[7,3] = -1j*gamma_2*k[2]/2

  # The Hamiltonian matrix is now fully defined and can be used for further calculations
  return H


def calc_band_diagram(points=20):
  band_diagram = np.zeros((8,points), dtype=np.float64)
  
  k_x = np.linspace(0,0.02,points)
  for i, k_x_i in enumerate(k_x):      
    H = eight_band_kp_Hamiltonian([k_x_i,0,0])
    evals, evects = np.linalg.eig(H)
    band_diagram[:,i] = evals.astype(np.float64)
    
  return band_diagram, k_x

bd, k_x = calc_band_diagram(100)

bd.sort(axis=0)
with plt.style.context('bmh'):
  plt.figure(figsize=(3,8))
  for band in range(8):
    plt.plot(k_x, bd[band], '-')
  plt.show()
```
